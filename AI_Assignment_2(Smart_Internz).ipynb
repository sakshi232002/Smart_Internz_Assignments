#**5. Check for Missing values and deal with them.**

```python
df.isna().value_counts()
```

```python
df['culmen_length_mm'].fillna(df['culmen_length_mm'].mean(),inplace=True)
df['culmen_depth_mm'].fillna(df['culmen_depth_mm'].mean(),inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean(),inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].mean(),inplace=True)
df['sex'].fillna(df['sex'].mode()[0],inplace=True)
```

```python
df["sex"].isnull()
```

```python
df.isna().value_counts()
```

#**7. Check for Categorical columns and perform encoding**

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['species']=le.fit_transform(df['species'])
df['island']=le.fit_transform(df['island'])
df['sex']=le.fit_transform(df['sex'])
```

```python
df.info()
```

#**8. Split the data into dependent and independent variables.**

```python
x=df.drop(['species'],axis=1)
y=df['species']
```

```python
y.shape
```

#**10. Split the data into training and testing**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
print("Training set X shape:", x_train.shape)
print("Training set y shape:", y_train.shape)
print("Testing set X shape:", x_test.shape)
print("Testing set y shape:", y_test.shape)
```

#**11. Build the Model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestClassifier
```

```python
df=DecisionTreeClassifier(criterion='entropy',max_depth=2)
lr=LinearRegression()
r=Ridge()
l=Lasso()
rf=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
```

#**12. Train the Model**

```python
df.fit(x_train,y_train)
```

```python
lr.fit(x_train,y_train)
```

```python
r.fit(x_train,y_train)
```

```python
l.fit(x_train,y_train)
```

```python
rf.fit(x_train,y_train)
```

#**13. Test the Model**

```python
pred1=df.predict(x_test)
pred2=lr.predict(x_test)
pred3=r.predict(x_test)
pred4=l.predict(x_test)
pred5=rf.predict(x_test)
```

# **14. Measure the performance using Evaluation Metrics.**

```python
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error,classification_report
```

```python
print(f"for Decision Tree Model \nMSE={mean_squared_error(y_test,pred1)} \nR2_score={r2_score(y_test,pred1)}")
print(f"for Linear Regression Model \nMSE={mean_squared_error(y_test,pred1)} \nR2_score={r2_score(y_test,pred2)}")
print(f"for Ridge Model \nMSE={mean_squared_error(y_test,pred1)} \nR2_score={r2_score(y_test,pred3)}")
print(f"for Lasso Model \nMSE={mean_squared_error(y_test,pred1)} \nR2_score={r2_score(y_test,pred4)}")
print(f"for Random Forest Classifier Model \nMSE={mean_squared_error(y_test,pred1)} \nR2_score={r2_score(y_test,pred5)}")
```
