**1. Write a python Program to display the last digit of number(hint: Any number divide by 10 will return last digit)**

```python
def last_num(number):
  return number % 10
number=int(input("Enter a number"))
print("The last digit of",number,"is",last_num(number))
```

**2. Write a function to Calculate Area and Perimeter of Rectangle**

```python
#Area of Rectangle
def area_rect(length,width):
  return length * width

#perimeter of rectangle
def peri_rect(length,width):
  return 2*length + width

length=float(input("Enter a length of rectangle : "))
width=float(input("Enter a width of rectangle : "))

area_rect=area_rect(length,width)
peri_rect=peri_rect(length,width)

print("Area of Rectangle is : ",area_rect)
print("Perimeter of Rectangle is : ",peri_rect)
```

**3. Write a python Program to find out given No is Prime or Not.**

```python
num=int(input("Enter a Number : "))
if num >1:
  for i in range(2,num):
    if (num % i)==0:
      print("It is not a prime number",num)
      break
    else:
      print("It is a prime number",num)
      break
else:
  print("It is not a prime number",num)
```

**5. Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price
First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)**

```python
Unit=int(input("Enter no of unit : "))
if Unit<= 100:
  print("No Charge")
elif Unit >=100 and Unit<=200:
  x=5*Unit
  print("Charge is",x)
else:
  y=10*Unit
  print("Charge is",y)
```
